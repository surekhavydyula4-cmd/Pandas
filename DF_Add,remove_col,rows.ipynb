```python
#Adding status column
import pandas as pd
df=pd.read_csv("sales8.csv")
df["Status"]="Delivered"
print(df.head())
```

```python
#Adding Total cost Column
import pandas as pd
df=pd.read_csv("sales8.csv")
df["Total"]=df["Product cost"]*df["Quantity"]
print(df.head())
```

```python
#Adding column at specific position
import pandas as pd
df=pd.read_csv("sales8.csv")
new=df["Product cost"]*df["Quantity"]
df.insert(4,"Total",new)
print(df.head())
```

```python
#Adding column using apply() method
import pandas as pd
df=pd.read_csv("sales8.csv")
def total(df):
  t=df["Product cost"]*df["Quantity"]
  return t
df["Total"]=df.apply(total,axis=1)
print(df.head())
```

```python
#Dropping columns
import pandas as pd
df=pd.read_csv("sales8.csv")
df1=df.drop(columns="Quantity")
print(df1.head())
```

```python
#Dropping columns
import pandas as pd
df=pd.read_csv("sales8.csv")
df1=df.drop(["Customer Id","Product cost"],axis=1)
print(df1.head())
```

```python
#Dropping rows
import pandas as pd
df=pd.read_csv("sales8.csv")
df1=df.drop(3,axis=0)
print(df1.head())
```

```python
#Dropping rows
import pandas as pd
df=pd.read_csv("sales8.csv")
df1=df.drop([0,5],axis=0)
print(df1.head())
```
