```python
#creating year csv files with all months files
import os
import glob
import pandas as pd
p='/content/Surekha'

files=os.path.join(p,'*.csv')
csv_files=glob.glob(files)

result=(pd.read_csv(csv)for csv in csv_files)

df=pd.concat(result,ignore_index=True)

print(df)

df.to_csv("Year.csv",index=False)
```
