```python
#sort_values()
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df.sort_values(by="Product_Cost")
print(df1.head())
```

```python
#sort_values()
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df.sort_values(by="Customer_Name")
print(df1.head())
```

```python
#sort_values() with ascending=False
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df.sort_values(by="Product_Name",ascending=False)
print(df1.head())
```

```python
#sort_values() with ascending=0
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df.sort_values(by="Order_Id",ascending=0)
print(df1.head())
```

```python
#sort_values() with ascending=0 without by
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df.sort_values("Order_Id",ascending=0)
print(df1.head())
```

```python
#sort_values() with multiple column anmes throw error
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=df.sort_values(by=["Order_Id","Product_Name"],ascending=0)
print(df1.head())
```

```python
#sort_index
import pandas as pd
d = {
        'Order id': [11, 21, 31],
        'Customer name': ['Kedar', 'Nireekshan', 'Daniel'],
        'Product': ['iPhone 11','hTC', 'macbook']
}

i = [555, 444, 333]
df=pd.DataFrame(d,index=i)
df1=df.sort_index()
print(df1)
```

```python
#sort_index with inplace parameter
import pandas as pd
d = {
        'Order id': [11, 21, 31],
        'Customer name': ['Kedar', 'Nireekshan', 'Daniel'],
        'Product': ['iPhone 11','hTC', 'macbook']
}
i = [555, 444, 333]
df=pd.DataFrame(d,index=i)
df.sort_index(inplace=True)
print(df)
```
