```python
#loc to access particular prodcuts and columns using slice
import pandas as pd
df=pd.read_csv("sales2.csv")
df.set_index("Product name",inplace=True)
a=["Flatscreen TV","Samsung Galaxy S20"]
df1=df.loc[a,"Order id":"Customer id"]
print(df1)
```
