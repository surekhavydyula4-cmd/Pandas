```python
#Finding missing values with isna()
import pandas as pd
df=pd.read_csv("fruits1.csv")
df1=df.isna()
print(df.head())
print()
print(df1.head())
```

```python
#Finding missing values with isnull()
import pandas as pd
df=pd.read_csv("fruits1.csv")
df1=df.isnull()
print(df.head())
print()
print(df1.head())
```

```python
#Count the number of missing values in each column
import pandas as pd
df=pd.read_csv("fruits1.csv")
df1=df.isna().sum()
print(df1)
```

```python
#Finding the percentage of missing values
import pandas as pd
df=pd.read_csv("fruits1.csv")
s=df.isna().sum()
df1=(s*100)/len(df)
print(df1)
```

```python
#Handle missing values with dropna()
import pandas as pd
df=pd.read_csv("fruits1.csv")
df1=df.dropna()
print(df)
print()
print(df1)
```

```python
#Handle missing values with dropna(inplace=True)
import pandas as pd
df=pd.read_csv("fruits1.csv")
df.dropna(inplace=True)
print(df)
```

```python
#Converting data type from float to int
import pandas as pd
df=pd.read_csv("fruits1.csv")
df1=df.dropna()
df2=df1.astype(int)
print(df1.head())
print()
print(df2.head())
```

```python
#Handle missing values with fillna(0)
import pandas as pd
df=pd.read_csv("fruits1.csv")
df1=df.fillna(0)
print(df)
print()
print(df1)
```

```python
#Handle missing values with fillna(number)
import pandas as pd
import numpy as np
d=[
    [10,"Sai",22,20000.0],
    [11,"Mega",np.nan,20000.0],
    [12,"Lucky",np.nan,30000.0],
    [14,"Sri",23,30000.0],
    [13,"Suri",22,25000.0],
    [15,"Cherry",21,30000.0]
]
df=pd.DataFrame(d,columns=["Id","Name","Age","Salary"])
df1=df.fillna(22)
print(df)
print()
print(df1)
```

```python
#Handle missing values with fillna(number)
import pandas as pd
import numpy as np
d=[
    [10,"Sai",22,20000.0],
    [11,"Mega",np.nan,20000.0],
    [12,"Lucky",np.nan,30000.0],
    [14,"Sri",23,30000.0],
    [13,"Suri",22,25000.0],
    [15,"Cherry",21,30000.0]
]
df=pd.DataFrame(d,columns=["Id","Name","Age","Salary"])
print(df)
print()
m=df["Age"].mean()
df["Age"]=df["Age"].fillna(m)
print(df)
```

```python
#Handle missing values with replace()
import pandas as pd
import numpy as np
df=pd.read_csv("fruits1.csv")
df1=df.replace(np.nan,0)
print(df)
print()
print(df1.head())
```
