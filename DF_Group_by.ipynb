```python
#Single function to columns inn group
import pandas as pd
df=pd.read_csv("sales5.csv")
d={
    "Product_Cost":sum
}
col=["Date","Product_Name"]
g=df.groupby(col)
r=g.agg(d)
print(r)
```

```python
#Single function to columns in group
import pandas as pd
df=pd.read_csv("sales5.csv")
d={
    "Product_Cost":sum,
    "Product_Name":"count"
}
col=["Date","Product_Name"]
g=df.groupby(col)
r=g.agg(d)
print(r)
```

```python
#Single function to columns in group
import pandas as pd
df=pd.read_csv("sales5.csv")
d={
    "Product_Cost":sum,
    "Product_Name":"count"
}
col=["Mail_Id","Date"]
g=df.groupby(col)
r=g.agg(d)
print(r)
```

```python
#Single function to columns in group
import pandas as pd
df=pd.read_csv("sales5.csv")
d={
    "Product_Cost":[min,max,sum]
}
col=["Date","Product_Name"]
g=df.groupby(col)
r=g.agg(d)
print(r)
```
