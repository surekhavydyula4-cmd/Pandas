```python
#Creating DataFrames
import pandas as pd
d1=[[11,22],[33,44]]
d2=[[55,66],[77,88]]

col=["COL1","COL2"]

df1=pd.DataFrame(d1,columns=col)
df2=pd.DataFrame(d2,columns=col)
print(df1)
print()
print(df2)
```

```python
#Concat two dataframes
import pandas as pd
d1=[[11,22],[33,44]]
d2=[[55,66],[77,88]]

col=["COL1","COL2"]

df1=pd.DataFrame(d1,columns=col)
df2=pd.DataFrame(d2,columns=col)

d=[df1,df2]
df3=pd.concat(d)
print(df3)
```

```python
#concat two dataframes ignore index
import pandas as pd
d1=[[11,22],[33,44]]
d2=[[55,66],[77,88]]

col1=["COL1","COL2"]
col2=["COL3","COL4"]

df1=pd.DataFrame(d1,columns=col1)
df2=pd.DataFrame(d2,columns=col2)

d=[df1,df2]
df3=pd.concat(d,ignore_index=True)

print(df3)
```

```python
#Concat DataFrames
import pandas as pd
d1=[[11,22],[33,44]]
d2=[[55,66],[77,88]]

col=["COL1","COL2"]

df1=pd.DataFrame(d1,columns=col)
df2=pd.DataFrame(d2,columns=col)

d=[df1,df2]
df3=pd.concat(d,axis=1)
print(df3)
```
