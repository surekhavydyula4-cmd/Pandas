```python
#fliter data using multiple conditions
import pandas as pd
df=pd.read_csv("sales4.csv")
a=df["Product_Name"]=="Samsung Galaxy S20"
b=df["Customer_Name"]=="Tarun"
df1=df[a&b]
print(df1)
```

```python
#filter data using multiple conditions
import pandas as pd
df=pd.read_csv("sales4.csv")
a=df["Order_Id"]==1050
b=df['Customer_Id']==10
df1=df[a|b]
print(df1)
```

```python
#filter using loc indexer
import pandas as pd
df=pd.read_csv("sales4.csv")
a=df["Customer_Id"]>24
df1=df.loc[a]
print(df1)
```

```python
#Filtering using loc indexer and multiple conditions
import pandas as pd
df=pd.read_csv("sales4.csv")
a=df["Customer_Name"]=="Tarun"
b=df["Order_Id"]>1640
df1=df.loc[a&b]
print(df1)
```

```python
#Filtering using loc indexer with multiple codtions
import pandas as pd
df=pd.read_csv("sales4.csv")
a=df["Product_Name"]=="iPhone 9"
b=df["Order_Id"]==1034
df1=df.loc[a|b]
print(df1)
```

```python
#Filtering with loc indexer and select with rows and ccolumns
import pandas as pd
df=pd.read_csv("sales4.csv")
r=df.index[0:]
c=["Customer_Id","Product_Name"]
df1=df.loc[r,c]
print(df1)
```

```python
#Filtering with loc indexer and select with rows and ccolumns
import pandas as pd
df=pd.read_csv("sales4.csv")
r=df.index[0:5]
c=["Customer_Name","Product_Name"]
df1=df.loc[r,c]
print(df1)
```

```python
#isin() method
import pandas as pd
df=pd.read_csv("sales4.csv")
a=["iPhone 11"]
b=df["Product_Name"].isin(a)
df1=df[b]
print(df1)

```

```python
#isin() method
import pandas as pd
df=pd.read_csv("sales4.csv")
a=["Chaithanya"]
b=df["Customer_Name"].isin(a)
df1=df[b]
print(df1)
```

```python
# isin() method
import pandas as pd
d={
    "Order_id":[10,11,12,13],
    "Customer_Name":["Sai","Cherry","Suri","Paddu"],
    "Product_Name":["iPhone 11","macbook","Samsung Galaxy","iPhone 9"]
}
df=pd.DataFrame(d)
a=["Sai"]
b=df.Customer_Name.isin(a)
df1=df[b]
print(df1)
```

```python
# unique() function
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=pd.unique(df["Customer_Name"])
print(df1)
```

```python
# unique() function
import pandas as pd
df=pd.read_csv("sales4.csv")
df1=pd.unique(df["Order_Id"])
print(df1)
```

```python
# notnull() method
import pandas as pd
d={
    "Order_id":[10,11,12,13],
    "Customer_Name":["Sai","Cherry","Suri","Paddu"],
    "Product_Name":["iPhone 11","macbook","Samsung Galaxy","iPhone 9"]
}
df=pd.DataFrame(d)
a=df.Product_Name.notnull()
df1=df[a]
print(df1)
```
